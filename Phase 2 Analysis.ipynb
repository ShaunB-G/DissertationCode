## Aggravated Offences vs hate tweet lag

```python
import numpy as np
from scipy import stats

results = []

# Define a function to handle correlations and p-values while ignoring NaNs
def calculate_corr_pval(x, y):
    mask = ~np.isnan(x) & ~np.isnan(y)
    return stats.pearsonr(x[mask], y[mask])


# Calculate lagged correlations and p-values for each shift
shifts = [-3, -2, -1, 0, 1, 2, 3, 4, 5, 6]  # Negative values for future correlations, positive for past correlations
for shift in shifts:
        UKmonthlyHate2015[f'Hate_Tweets_Lagged_{shift}'] = UKmonthlyHate2015['Hate_Tweet_Percentage'].shift(shift)
        correlation, p_value = calculate_corr_pval(UKmonthlyHate2015[f'Hate_Tweets_Lagged_{shift}'], UKmonthlyHate2015['Aggravated offences'])
        print(f"Lagged correlation (Aggravated offences correlation with Hate Tweets {shift} months in the past): {correlation}, p-value: {p_value}")
        results.append({'Shift': shift, 'Correlation': correlation, 'P-Value': p_value})
```

```python
# Convert results to a DataFrame for tabular display
results_df = pd.DataFrame(results)

# Display the DataFrame (the table of results)
print(results_df)

# Create a line graph highlighting statistically significant results (p-value < 0.05)
plt.figure(figsize=(10, 6))
plt.plot(results_df['Shift'], results_df['Correlation'], label='Correlation', marker='o', color='blue')

# Highlight statistically significant points
significant = results_df['P-Value'] < 0.05
plt.scatter(results_df['Shift'][significant], results_df['Correlation'][significant], color='red', label='p < 0.05', zorder=5)

# Add labels and title
plt.axhline(0, color='black', linewidth=0.5)
plt.xlabel('Hate Tweets Shift (Months)')
plt.ylabel('Correlation Coefficient')
plt.title('Lagged Correlation between Hateful Tweets and Aggravated Offences')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from scipy.stats import ttest_ind, wilcoxon

# Separate pre- and post-Brexit data
pre_brexit = BrexitVotedf[BrexitVotedf.index < 'June 2016']
post_brexit = BrexitVotedf[BrexitVotedf.index >= 'June 2016']

# Perform t-tests
t_test_results = {}
for column in ['Hate_Tweet_Percentage', 'Total Offences', 'Non-aggravated offences', 'Aggravated offences']:
    t_stat, p_val = ttest_ind(pre_brexit[column], post_brexit[column])
    t_test_results[column] = {'t-statistic': t_stat, 'p-value': p_val}

# Optionally, use Wilcoxon signed-rank test if normality is an issue
wilcoxon_results = {}
for column in ['Hate_Tweet_Percentage', 'Total Offences', 'Non-aggravated offences', 'Aggravated offences']:
     w_stat, p_val = wilcoxon(pre_brexit[column], post_brexit[column])
     wilcoxon_results[column] = {'Wilcoxon statistic': w_stat, 'p-value': p_val}

# View results
t_test_results
# wilcoxon_results  # Uncomment if using Wilcoxon instead
```

```python
import statsmodels.api as sm
import numpy as np

# Add a time variable for each month
BrexitVotedf['Time'] = np.arange(len(BrexitVotedf))

# Add an intervention variable (0 for pre-Brexit, 1 for post-Brexit)
BrexitVotedf['Intervention'] = (BrexitVotedf.index >= '2016-06-01').astype(int)

# Add a time-after-intervention variable (0 for pre-Brexit months, increasing post-Brexit)
BrexitVotedf['Time_after_Intervention'] = BrexitVotedf['Time'] * BrexitVotedf['Intervention']

# Define the model for Hate Tweet Percentage
X = BrexitVotedf[['Time', 'Intervention', 'Time_after_Intervention']]
X = sm.add_constant(X)  # Adds an intercept
y = BrexitVotedf['Hate_Tweet_Percentage']

# Fit the model
model = sm.OLS(y, X).fit()
print(model.summary())

# Repeat for 'Total Offences'
y_offences = BrexitVotedf['Total Offences']
model_offences = sm.OLS(y_offences, X).fit()
print(model_offences.summary())

# Repeat for 'Aggravated Offences'
y_offences = BrexitVotedf['Aggravated offences']
model_offences = sm.OLS(y_offences, X).fit()
print(model_offences.summary())

# Repeat for 'Non-Aggravated Offences'
y_offences = BrexitVotedf['Non-aggravated offences']
model_offences = sm.OLS(y_offences, X).fit()
print(model_offences.summary())
```
